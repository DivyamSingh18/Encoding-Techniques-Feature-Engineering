# Probability Ratio Encoding
1. Probability of Survived based on Cabin--- Categorical Feature
2. Probability of Not Survived---1-pr(Survived)
3. pr(Survived)/pr(Not Survived)
4. Dictonary to map cabin with probability
5. replace with the categorical feature

```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
df.head()
```

```python
### Replacing NaN with Missing
df['Cabin'].fillna('Missing',inplace=True)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]  # taking the first letter as category 
df
```

```python
df.Cabin.unique()
```

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean() # finding mean of survived w.r.t. Cabin
prob_df
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died']= 1 - prob_df['Survived']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio']= prob_df['Survived']/prob_df['Died'] #finding probability raito
prob_df.head()
```

```python
probability_encoded = prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded']=df['Cabin'].map(probability_encoded)
df.head()
```

# Advantages of probability ratio encoding
1. Captures information within the category, and therefore creates more predictive features.
2. Creates a monotonic relationship between the variables and the target. So itâ€™s suitable for linear models.
3. Does not expand the feature space.

# Limitations of probability ratio encoding
1. Likely to cause overfitting.
2. Not defined when the denominator is 0.
