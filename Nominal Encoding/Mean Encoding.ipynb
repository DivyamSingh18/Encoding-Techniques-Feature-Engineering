# Mean Encoding
we calculate the mean for each category and replace the variable with the mean value.

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv', usecols=["Cabin","Survived"])
df
```

```python
df['Cabin'].fillna('Missing',inplace=True) #replacing NaN values with Missing
df
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0] # taking first lettters from strings in Cabin
df
```

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
df['mean_encode']=df['Cabin'].map(mean_ordinal)
df
```
